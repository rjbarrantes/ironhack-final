```python
import warnings
warnings.simplefilter(action = 'ignore', category = FutureWarning)
warnings.simplefilter(action = 'ignore', category = DeprecationWarning)

import keras
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, BatchNormalization
from keras.activations import relu, softmax, sigmoid
from keras.optimizers import Adam, SGD
from keras.losses import binary_crossentropy, categorical_crossentropy, sparse_categorical_crossentropy
from keras.metrics import binary_accuracy, categorical_accuracy, sparse_categorical_accuracy
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = .8)
```

```python
model0 = Sequential()

model0.add(Conv2D(16, (3, 3), activation = relu, input_shape = X_train[0].shape))
model0.add(BatchNormalization())
model0.add(MaxPooling2D())
model0.add(Dropout(.2))

model0.add(Conv2D(32, (3, 3), activation = relu))
model0.add(BatchNormalization())
model0.add(MaxPooling2D())
model0.add(Dropout(.3))

model0.add(Conv2D(64, (3, 3), activation = relu))
model0.add(BatchNormalization())
model0.add(MaxPooling2D())
model0.add(Dropout(.4))

model0.add(Conv2D(128, (3, 3), activation = relu))
model0.add(BatchNormalization())
model0.add(MaxPooling2D())
model0.add(Dropout(.5))

model0.add(Flatten())

model0.add(Dense(128, activation = relu))
model0.add(BatchNormalization())
model0.add(Dropout(.5))

model0.add(Dense(4, activation = sigmoid))

model0.summary()
```

```python
model00 = model0.compile(loss = binary_crossentropy, optimizer = Adam(), metrics = ['accuracy'])
history0 = model00.fit(X_train, y_train, epochs = 10, validation_data = (X_test, y_test))
```
