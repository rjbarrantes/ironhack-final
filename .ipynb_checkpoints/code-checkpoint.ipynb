```python
# read image files

import pathlib
import imageio
import numpy as np
from skimage.color import rgb2gray

training_images = [i for i in pathlib.Path('../../data/understanding_cloud_organization/train_images').glob('*.jpg')]
im = imageio.imread(str(training_images[1]))
im_gray = rgb2gray(im)
```

```python
# compare shapes

print(f'original shape = {im.shape}')
print(f'grayscale shape = {im_gray.shape}')
```

```python
# compare images

import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(im)
plt.axis('off')
plt.title('original')

plt.subplot(1,2,2)
plt.imshow(im_gray, cmap='gray')
plt.axis('off')
plt.title('grayscale')

plt.tight_layout()
plt.show()
```
